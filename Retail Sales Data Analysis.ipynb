```python
import numpy as np
```

```python
#plot between min and max with unit steps
x_axis = np.arange(df['single_price'].min(),df['single_price'].max())
print(x_axis)
```

```python
import statistics
#calculating mean and std
mean = statistics.mean(x_axis)
std = statistics.stdev(x_axis)
print("Mean",mean)
print("Standard Deviation",std)
```

```python
from scipy.stats import norm
plt.plot(x_axis,norm.pdf(x_axis,mean,std))
plt.title("Normal Distribution od price attribute")
```

```python
x_axis1 = np.arange(df['amount'].min(),df['amount'].max())
print(x_axis1)
```

```python
import numpy as np
#calculating mean and std
mean = np.mean(x_axis1)
std = np.std(x_axis1)
print("Mean",mean)
print("Standard Deviation",std)
```

```python
from scipy.stats import norm
plt.plot(x_axis1,norm.pdf(x_axis1,mean,std))
plt.title("Normal Distribution of amount attribute")
```
